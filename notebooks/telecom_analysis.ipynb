# Telecom Customer Churn Analysis

## Exploratory Data Analytics on Customer Billing, Service Usage, and Churn Behavior

### Project Objective

The purpose of this analysis is to explore telecom customer behavior and identify patterns associated with customer churn.

This notebook investigates:
- the overall churn distribution,
- the relationship between customer contracts and churn,
- the impact of monthly billing on churn,
- and other service-related factors that may influence customer retention.

### Importing Required Libraries

The following libraries are imported for data manipulation, visualization, and exploratory analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Dataset

The telecom customer churn dataset is loaded directly from the GitHub repository into the Google Colab environment for analysis.

```python

url = 'https://raw.githubusercontent.com/Azure-priscy/telecom-churn-analysis-project/refs/heads/main/data/Telecom%20Customers%20Churn.csv'
df = pd.read_csv(url)

df.head()
```

### Churn Distribution Analysis

The first analytical step is to determine the proportion of customers who stayed versus customers who left the telecom company.

```python
df["Churn"].value_counts().plot(kind="bar")
```

### Contract Type and Customer Churn

This section investigates whether customer contract duration has any visible relationship with churn behavior.

```python
import seaborn as sns

sns.countplot(data=df, x="Contract", hue="Churn")
plt.title("Churn by Contract Type")
plt.show()
```

### Monthly Charges and Churn Relationship

This section examines whether customers paying higher monthly charges are more likely to churn.

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```
